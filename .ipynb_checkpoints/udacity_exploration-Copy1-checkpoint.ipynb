```python
# get year from datetime: 1. define a function to extract the year
def extract_year(data_string):
    date_object = pd.to_datetime(data_string)
    return date_object.year

df_filter['Year'] = df_filter['ListingCreationDate'].apply(extract_year)

print(df_filter['Year'])
```

```python
def status(value):
    if value ==  "Chargedoff" or value == "Defaulted" :
        return "Loss"
    elif value == "Past Due (1-15 days)" or value == "Past Due (31-60 days)" or value == "Past Due (91-120 days)" or value == "Past Due (61-90 days)" or value == "Past Due (16-30 days)" or value == "Past Due (>120 days)":
        return "Due"
    elif value == "Current":
        return "Current"
    elif value == "Completed":
        return "Completed"
    

df_filter['StatusLoan'] = df_filter['LoanStatus'].map(status)

print(df_filter)
```

You can use LabelEncoder.classes_ and LabelEncoder.transform() to get the relationships you're asking for. The following function should give you what you need.

def get_integer_mapping(le):
    '''
    Return a dict mapping labels to their integer values
    from an SKlearn LabelEncoder
    le = a fitted SKlearn LabelEncoder
    '''
    res = {}
    for cl in le.classes_:
        res.update({cl:le.transform([cl])[0]})

    return res
    
Example Usage:
classes = ['blue', 'green']
le = LabelEncoder()
le.fit(classes)
integerMapping = get_integer_mapping(le)

integerMapping['blue']  # Returns 0
integerMapping['green']  # Returns 1
If you have a very large number of classes, this may not be terribly fast due to the for loop, but it should work well for a small number of classes.

```python


df_filter['AvailableBankcardCredit'].hist(bins=150,grid = False)

# plot histogram using quartile
# - compute quartile lines
q1, q2, q3 = np.percentile(df_filter['AvailableBankcardCredit'], [25, 50, 75])

# Plot histogram using quartile
fig, ax = plt.subplots()
ax.hist(df_filter['AvailableBankcardCredit'], bins=20)
ax.axvline(x=q1, color='red', linestyle='--', label='Q1')
ax.axvline(x=q2, color='green', linestyle='--', label='Q2')
ax.axvline(x=q3, color='blue', linestyle='--', label='Q3')
ax.legend()

# Add labels and title
ax.set_xlabel('Available Bankcard Credit')
ax.set_ylabel('Frequency')
ax.set_title('Histogram with Quartiles')

# Show the plot
plt.show()
```

```python
# Compute cutoff values
q01, q99 = df_filter['AvailableBankcardCredit'].quantile([0.01, 0.99])

# Filter out values outside the cutoff range
df_filtered = df_filter[(df_filter['AvailableBankcardCredit'] >= q01) & (df_filter['AvailableBankcardCredit'] <= q99)]

# Quartile from df_filtered
q1, q2, q3 = np.percentile(df_filtered['AvailableBankcardCredit'], [25, 50, 75])

# Plot histogram of filtered data
fig, ax = plt.subplots()
ax.hist(df_filtered['AvailableBankcardCredit'], bins=100)
ax.axvline(x=q1, color='red', linestyle='--', label='Q1')
ax.axvline(x=q2, color='green', linestyle='--', label='Q2')
ax.axvline(x=q3, color='blue', linestyle='--', label='Q3')
ax.legend()

# Add labels and title
ax.set_xlabel('Available Bankcard Credit')
ax.set_ylabel('Frequency')
ax.set_title('Histogram with 1% Cutoff')

# Show the plot
plt.show()
```
